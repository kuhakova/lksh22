# Практическое занятие по спектроскопии поверхности Марса

## Введение

В этом руководстве мы собрали описания и типовые местоположения некоторых уникальных спектральных особенностей, выявленных на планете Марс с использованием данных прибора CRISM. 
Это задание предназначено в качестве ознакомления с тем, что ученые-планетологи обнаружили на данный момент, и вкратце показывает, как однозначно идентифицировать минералы в различных местах. Этот документ содержит расширенную и несколько менее техническую версию информации, представленной в научных публикации [Viviano-Beck et al. (2014)](https://doi.org/10.1002/2014JE004627) по спектроскопии Марса.  


На графических материалах отображено название места и географические координаты. Приводятся идентификатор наблюдения CRISM (например, **FRT0000C26C**) для типовой местности, год и день года, когда было получено наблюдение (например, 2008_235 означает 235 день 2008-го года), и время года, когда было проведено наблюдение. Время года выражено в виде [ареоцентрической долготы Марса](https://ru.wikipedia.org/wiki/%D0%A5%D1%80%D0%BE%D0%BD%D0%BE%D0%BC%D0%B5%D1%82%D1%80%D0%B8%D1%8F_%D0%BD%D0%B0_%D0%9C%D0%B0%D1%80%D1%81%D0%B5) (L$_s$). L$_s$ колеблется от 0 до 360 градусов; 0° — весеннее равноденствие в северном полушарии, 90° — летнее солнцестояние в северном полушарии, 180° — осеннее равноденствие в северном полушарии и 270° — день весеннего равноденствия в северном полушарии.


В этой директории находится три папки с данными прибора CRISM. В каждой из папок находится картинка с двумя изображениям. На первом (левом) изображении мы видим наблюдения CRISM в видимом цвете (т. е. с преувеличенными цветовыми различиями) и на втором &mdash; обзорные продукты, состоящие из сводных параметров, которые выделяют спектральные вариации на поверхности Марса, связанные с минералом, представляющим интерес. Продукт просмотра &mdash; синтезированное изображение, в котором каждому цветовому каналу в красно-зелено-синем формате присвоен сводный параметр (например, глубина полосы или какой-либо другой расчетный показатель спектральной изменчивости). Читатель может обратиться к следующей таблице для более подробного объяснения каждого продукта. 
![](table.png)

Линейка продуктов и сводных параметров гораздо шире, чем представлена в данном задании. Для некоторых отдельных минералов мы используем стандартные продукты просмотра, чтобы выделить интересующий нас минерал; однако иногда используется специальная композиция, чтобы интересующий минерал был более заметен или отличался от других минералов на типовой местности. 


Сводный параметр означает поглощение в спектре, то есть глубину полосы («BD» &mdash; Band Depth) например, `BD2900` – поглощение на длине волны 2.9 микрон. В продуктах обзора CRISM каждому цветовому каналу назначается один суммарный параметр. Например, продукт просмотра PHY представляет собой набор параметров `D2300`, `D2200` и `BD1900_2`, каждый из которых назначен каналам RGB соответственно. Если бы только один параметр имел повышенные значения в изображении, то вы бы наблюдали пиксели с цветом, связанным с этим параметром, например, красные пиксели, соответствующие `D2300`. Однако минерал обычно имеет более одного поглощения, поэтому более одного параметра могут иметь повышенные значения для данного пикселя. Два высоких значения параметра приведут к голубому, пурпурному или желтому цвету. Таким образом, пиксели, которые выглядят пурпурными на изображении PHY, имеют повышенные значения `D2300` и `BD1900_2`, которые обычно коррелируют со смектитами Fe/Mg. Точно так же пиксели, которые выглядят голубыми, представляют собой комбинацию `D2200` и `BD1900_2`. Белые пиксели указывают, что все три параметра имеют высокие значения, а черные указывают на то, что ни один из параметров не имеет высоких значений.


После изучения полученной карты с продуктами обзора выбирается интересующее место, из которого извлекается информация со спектром. Данные спектры уже загружены в данную директорию. После этого по виду кривой определяется наиболее подходящий лабораторный спектр минерала. Сопоставив два спектра, можно достоверно говорить о том, что мы отождествили присутствие данного минерала в спектре CRISM в исследуемой локации. 

## Локация первая: HRL0000CA5C

На первом изображении в реальных цветах изображена исследуемая локация в [Olympia Undae](https://en.wikipedia.org/wiki/Olympia_Undae). Справа подобран продукт просмотра, который выделяет особенности минералогии данной местности. По подобранным цветам мы определяем наличие гипса. 
![](HRL0000CA5C/HRL0000CA5C.png)

Теперь необходимо убедиться в этой догадке с помощью спектрального анализа. После построения двух спектров мы должны сравнить полосы поглощения и подтвердить или опровергнуть сходство данных спектральных характеристик.


В этой директории есть папка `HRL0000CA5C`. Загрузите из нее спектр локации:

```python
from lksh import load_data
from lksh import plot
```

```python
wavelength, r = load_data('HRL0000CA5C/spectrum.txt') # впишите название файла
```

Теперь постройте:

```python
plot(wavelength, r, xlabel='Длина волны, мкм', ylabel='Коэффициент отражения') # впишите правильные аргументы
       # не забудьте подписать оси
       # описание данных есть в файле
```

Загрузите спектр гипса из этой же директории и постройте его:

```python
wavelength_gypsum, r_gypsum = load_data('HRL0000CA5C/gypsum.txt') # впишите название файла

plot(wavelength_gypsum, r_gypsum, xlabel='Длина волны, мкм', ylabel='Коэффициент отражения') 
       # впишите правильные аргументы
       # не забудьте подписать оси
       # описание данных есть в файле
```

- Определите соответствие лабораторного минерала и спектра локации

```python
plot(wavelength, r, xlabel='Длина волны, мкм', ylabel='Коэффициент отражения', label='спектр локации')
plot(wavelength_gypsum, r_gypsum, label='спектр гипса') 
```

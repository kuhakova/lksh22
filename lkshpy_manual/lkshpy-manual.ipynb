## Базовые функции и обращение с модулем

Для того чтобы использовать функцию из модуля, нам нужно загрузить этот модуль и импортировать из него функцию. `python` старается быть похожим на обычный английский, поэтому выглядит это так:

```python
from lksh import load_data
```

Мы загрузили "Из (модуля) `lksh` (функцию) `load_data`". Не забудьте нажать кнопку **Run** наверху выделив ячейку с кодом! А еще лучше нажать на ячейку с кодом и зажать кнопки `ctrl` + `enter`! \
\
Функция `load_data` загружает данные из .txt файла. Данные должны быть в двух или более колонках без какого-либо текста! Пример такого файла находится в `space-school-labs/astrophysics/example_data/sine.txt`. Его можно открыть в блокноте и посмотреть как он выглядит. Теперь давайте загрузим его! 

```python
x, y = load_data('example_data/sine.txt')
```

```python
x
```

*Обратите внимание: в скобочках мы указали полный путь, включая папку в которой находится файл. Этот путь считается от папки в который находится этот (`lksh-tool-manual.ipynb`) файл. Можно указать полный путь, начиная от диска `C:\`, но он может быть уникальный для каждого компьютера.*
\
\
Теперь у нас есть данные в двух переменных x и y. Давайте посмотрим что находится в этих переменных:

```python
print(x, y)
```

Числа являются далекой от идеальной репрезенантации данных. Давайте попробуем нарисовать график! Сначала импортируем функцию: 

```python
from lksh import plot
```

Теперь давайте построим данные:

```python
plot(x, y)
```

Теперь давайте подпишем оси и линию:

```python
plot(x, y, xlabel='Ось x', ylabel='Ось y', label='Синусоида')
```

Давайте попробуем загрузить другой файл. Он называется `example2.txt`:

```python
x, y = load_data('example_data/example2.txt')
```

Почему мы получили ошибку и что случилось? Если мы обратим внимание на последнюю строчку в файле, то увидим `ValueError: could not convert string to float: 'Пример'`. Давайте посмотрим как выглядит сам файл:

```python
cat 'example_data/example2.txt' # хитрый способ посмотреть содержимое файла
```

В данных есть не только два столбца чисел, но и какие-то другие строчки. Часто бывает, что данные бывают с какими-то дополнениями и комменатриями. К счастью, в `python` как и в других языках есть специальный символ `#` после которого все остальные дополнения считаются комменатрием к коду. Это же работает и для данных. Если вы откроете файл и закомментируете первые три строчки, например так:

```python
cat 'example_data/example_new.txt'
```

То он отлично загрузится и его можно будет нарисовать: 

```python
new_x, new_y = load_data('example_data/example_new.txt')
plot(new_x, new_y)
```

Теперь давайте наложим один график на другой:

```python
plot(x, y, xlabel='Ось x', ylabel='Ось y', label='Синусоида')
plot(new_x, new_y, label='Другой график')
```

**Примечание**: `xlabel` и `ylabel` достаточно указать только для одного графика, а вот `label` нужно указывать для каждого графика.
